```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
jugador_des= pd.read_csv("player_ratings.csv")
jugador_des.head()
```

```python
jugadores_test=pd.read_csv("test_player.csv")
jugadores_test.head()
```

```python
jugador_historial = pd.merge(jugadores_test,jugador_des, on = 'account_id',how = 'left')
jugador_historial.set_index(['match_id','player_slot'],inplace = True)
dim=len(jugador_historial)
jugador_historial.head()
```

```python
jugador_historial['ratio_v']=jugador_historial.apply(lambda row: row.total_wins/row.total_matches, axis=1)
```

```python
jugador_historial.head(20)
```

```python
jugadores_match=jugador_historial['ratio_v'].unstack()
jugadores_match.head()
```

```python
jugadores_match.dropna(inplace=True)
jugadores_match.head(20)
```

```python
#jugadores_match.columns.values
jugadores_match=jugadores_match[jugadores_match[:]>0.2]
jugadores_match.head()
```

```python
jugadores_match=jugadores_match[jugadores_match[:]<0.8]
jugadores_match.head()
```

```python
jugadores_match.dropna(inplace=True)
jugadores_match.head(20)
```

```python
partido_equipo=pd.read_csv("test_labels.csv")
partido_equipo.head()
```

```python
test_ratio=pd.merge(partido_equipo,jugadores_match,left_index = True,right_index = True)
test_ratio.head(20)
```

```python
test_ratio.isna().sum()
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

from sklearn.model_selection import cross_val_score
```

```python
X=test_ratio[test_ratio.columns.difference(["match_id","radiant_win"])]
y=test_ratio["radiant_win"]
y.head()
#list(test_pick)
X.head()
```

```python
X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(X,y,test_size=0.2)

print(X_entrenamiento)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
LR =LogisticRegression().fit(X_entrenamiento,y_entrenamiento)

predicciones_test_LR = LR.predict(X_prueba)

matriz_confusion_LR = metrics.confusion_matrix(y_prueba, predicciones_test_LR)
TN_LR = matriz_confusion_LR[0,0]
FN_LR = matriz_confusion_LR[1,0]
FP_LR = matriz_confusion_LR[0,1]
TP_LR = matriz_confusion_LR[1,1]

print ('              +-----------------+')
print ('              |   Predicción    |')
print ('              +-----------------+')
print ('              |    +   |    -   |')
print ('+-------+-----+--------+--------+')
print ('| Valor |  +  |   %d |  %d     |'   % (TP_LR, FN_LR) )
print ('| real  +-----+--------+--------+')
print ('|       |  -  | %d   | %d      |'    % (FP_LR, TN_LR) )
print ('+-------+-----+--------+--------+')
print ()
#print ( 'Exactitud    : ', (TP_LR+TN_LR)/(TP_LR+FN_LR+FP_LR+TN_LR) )
print('Exactitud     :',metrics.accuracy_score(y_prueba, predicciones_test_LR) )
print('Precisión     :',metrics.precision_score(y_prueba, predicciones_test_LR)) 
print('Exhaustividad :',metrics.recall_score(y_prueba, predicciones_test_LR))
print('F1-Score      :',metrics.f1_score(y_prueba, predicciones_test_LR))
```
